Q1. Explain what version control is and its importance in software development

Ans: Version control is a system that tracks changes to code over time. It's essential in software development because it:

  Allows multiple developers to collaborate safely.

   Lets you revert to previous versions if something goes wrong.

   Keeps a history of all changes made.

   Supports branching to develop features or fix bugs independently.

Tools like Git help manage code efficiently, prevent data loss, and improve teamwork.
